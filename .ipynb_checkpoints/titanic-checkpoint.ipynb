```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# import seaborn as sns
%matplotlib inline
```

```python
titanic=pd.read_csv("titanic-data.csv")
```

```python
titanic.head(3)
```

```python
len(titanic)
```

```python
titanic[["Age", "Fare", "Survived", "SibSp", "Pclass", "Parch"]].hist(figsize=(7,8))
plt.show()
```

```python
survived = titanic[titanic["Survived"]==1]
non_survived = titanic[titanic["Survived"]==0]
print("Column Names :", titanic.columns)
print("\nSurvived: ",len(survived))
print("\nNon Survived: ",len(non_survived))
print("\nRaw survival Rate: ",len(survived)/len(non_survived))
```

```python
fig=plt.figure(figsize=(6,4))
plt.subplot(221)
survived["Age"].hist(color="blue")
plt.subplot(222)
non_survived["Age"].hist(color="green")
plt.subplot(223)
survived["Age"].plot(kind="density", color="blue")
plt.subplot(224)
non_survived["Age"].plot(kind="density", color="green")
plt.tight_layout()
```

```python
titanic.boxplot(column="Age", by= "Survived", grid=False)
plt.show()
```

```python
# Based on Survival rate
plt.figure(figsize=(10,2))
plt.subplot(131)
titanic.groupby("Survived")["Pclass"].hist(alpha=0.5)
plt.subplot(132)
titanic.groupby("Survived")["Age"].hist(alpha=0.5)
plt.subplot(133)
titanic.groupby("Survived")["Fare"].hist(alpha=0.5)
# plt.subplot(144)
# titanic.groupby("Sex")["Survived"].hist(alpha=0.5)
# groups= titanic.groupby("Sex")["Survived"].hist()
# for k, v in groups:
#     v.hist(label=k, alpha=.5)

plt.show()
# plt.legend()
```

```python
# Survival rate and fare
plt.figure(1)
non_survived["Fare"].hist(color="green", alpha=0.5)
survived["Fare"].hist(color="blue", alpha=0.5)
plt.figure(2)
plt.subplot(221)
survived["Fare"].hist(color="blue")
plt.subplot(222)
non_survived["Fare"].hist(color="green")
plt.subplot(223)
survived["Fare"].plot(kind="density", color="blue")
plt.subplot(224)
non_survived["Fare"].plot(kind="density", color="green")
plt.figure(3)
titanic.boxplot(column = "Fare", by = "Survived", grid=False)
plt.ylim(-10, 250)
plt.show()
```

```python
# Based on Sex
plt.figure(figsize=(8,2))
plt.subplot(131)
titanic.groupby("Sex")["Pclass"].hist(alpha=0.5)
plt.subplot(132)
titanic.groupby("Sex")["Age"].hist(alpha=0.5)
plt.subplot(133)
titanic.groupby("Sex")["Fare"].hist(alpha=0.5)
plt.show()
```

```python
# Based on Survival rate
plt.figure(figsize=(8,2))
plt.subplot(131)
titanic.groupby("Survived")["Pclass"].hist(alpha=0.5)
plt.subplot(132)
titanic.groupby("Survived")["Age"].hist(alpha=0.5)
plt.subplot(133)
titanic.groupby("Survived")["Fare"].hist(alpha=0.5)
plt.show()
```
